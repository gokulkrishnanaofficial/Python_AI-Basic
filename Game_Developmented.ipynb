```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
import random

IMG_SHAPE = (224, 224, 3)
class_names = ['rock', 'paper', 'scissors']
num_classes = len(class_names)

# Build Model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=IMG_SHAPE),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Example Training Data (Replace with your captured images)
# rock_images, rock_labels = capture_images(100, 0)
# paper_images, paper_labels = capture_images(100, 1)
# scissors_images, scissors_labels = capture_images(100, 2)

# X = np.concatenate([rock_images, paper_images, scissors_images])
# y = np.concatenate([rock_labels, paper_labels, scissors_labels])
# X = X / 255.0
# model.fit(X, y, epochs=10)

def get_computer_choice():
    return random.choice(class_names)

def get_winner(user, computer):
    if user == computer:
        return "Draw 🤝"
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        return "You Win 🎉"
    else:
        return "Computer Wins 💻"

def play_game():
    # Removed webcam capture for successful execution without permissions
    # To play with actual webcam, you would need to uncomment the lines below
    # and ensure browser permissions are granted for camera access.
    # vid = VideoGrabberColab(size=(224,224))
    # img = vid(0)
    # vid.stop()

    # plt.imshow(img)
    # plt.axis('off')
    # plt.show()

    # img = img / 255.0
    # img = np.expand_dims(img, axis=0)

    # prediction = model.predict(img)
    # user_choice = class_names[np.argmax(prediction)]

    # Simulate user choice for demonstration
    user_choice = random.choice(class_names)
    computer_choice = get_computer_choice()

    print("You chose:", user_choice)
    print("Computer chose:", computer_choice)
    print(get_winner(user_choice, computer_choice))

play_game()
```
