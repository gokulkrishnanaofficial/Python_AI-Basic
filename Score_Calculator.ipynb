Import MODULES

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

Load Data

```python
data = pd.read_csv('student_marks.csv', sep=',')
print(data.head())
```

Separating Data

```python
predict = 'final' #output = label

x = np.array(data.drop(columns=[predict])) #input = feature
y = np.array(data[predict]) #Output = label

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.1)
```

Defining Model

```python
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test)
print(acc)
```

Predicting for specific student

```python
predictions = linear.predict(x_test)
for x in range(len(predictions)):
  print(f'Predicted value:{predictions[x]:0.1f}\t inputs: {x_test[x]}\t actual output:{y_test[x]}')
```
