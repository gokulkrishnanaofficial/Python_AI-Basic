```python
"""Class - Class is a set of functions(def) and attributes (more at)
__init__ initis the constructor method (function in class) for a class and it runs every time you create any object of the class.
It is run automatically every time so that it can define the attributes of the class objects with the method arguments
The self parameter refers to the attributes of the objects of the class"""
class robot:
  def __init__(self,n,c,w):
        self.name=n
        self.color=c
        self.weight=w
  def introduce_youself(self):
      print("my name is"+ self.name)
r1 = robot("Tom","red","30")
r2 = robot("Jerry","blue","40")
r1.introduce_youself()
r2.introduce_youself()

class flower:
  def __init__(self,n,c,a,s):
        self.name=n
        self.color=c
        self.area=a
        self.size=s
  def tell_property(self):
      print("my name is "+ self.name," color is "+self.color," area is "+self.area," size is "+self.size)
f1 = flower ("rose","red","30","L")
f2 = flower ("lilly","white","40","S")
f1.tell_property()
f2.tell_property()

"""Class Definition:
class flower: creates a blueprint for flower objects.

Constructor (__init__):
Runs automatically when a new object is created.
Initializes attributes: name, color, area, and size.

Method (tell_property):
Prints the details of the flower object in a descriptive sentence.
Uses the stored attributes to form the output.

Object Creation:
f1 = flower("rose","red","30","L") → creates a rose object.
f2 = flower("lilly","white","40","S") → creates a lilly object.

Method Calls:
f1.tell_property() → prints rose details.
f2.tell_property() → prints lilly details."""
```

```python
#chapter 1
'''Regression
The term regression is used when you try to find the relationship between variables.In Machine Learning, and in statistical modeling, that relationship is used to predict the outcome of future events.

*Linear Regression
*Polynomial Regression
*Multiple Regression

1.Linear Regression
Linear regression uses the relationship between the data-points to draw a straight line through all them. This line can be used to predict future values.

2.Polynomial Regression
If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression. Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.

3.Multiple Regression
Multiple regression is like linear regression, but with more than one independent value, meaning that we try to predict a value based on two or more variables.'''
```

```python
#Chapter 2
'''Classfification
Classification is a supervised machine learning technique used to predict labels or categories based on input data. The goal is to assign each data point to a predefined class, such as spam vs. non-spam emails or diseased vs. healthy patients.
For example, a classification model might be trained on dataset of images labeled as either dogs or cats and it can be used to predict the class of new and unseen images as dogs or cats based on their features such as colour, texture or shape.

Types of Classification
1. Binary Classification
2. Multiclass Classification
3. Multi-Label Classification

1.Binary Classification
This is the simplest kind of classification. In binary classification, the goal is to sort the data into two distinct categories. Think of it like a simple choice between two options. Imagine a system that sorts emails into either spam or not spam. It works by looking at different features of the email like certain keywords or sender details and decides whether it’s spam or not. It only chooses between these two options.

2.Multiclass Classification
Here, instead of just two categories, the data needs to be sorted into more than two categories. The model picks the one that best matches the input. Think of an image recognition system that sorts pictures of animals into categories like cat, dog and bird.

3. Multi-Label Classification
In multi-label classification single piece of data can belong to multiple categories at once. Unlike multiclass classification where each data point belongs to only one class, multi-label classification allows datapoints to belong to multiple classes. A movie recommendation system could tag a movie as both action and comedy. The system checks various features (like movie plot, actors or genre tags) and assigns multiple labels to a single piece of data, rather than just one.'''
```

```python
#Chapter 3
'''Clustering
Clustering is an unsupervised machine learning technique that groups similar data points together into clusters based on their characteristics, without using any labeled data. The objective is to ensure that data points within the same cluster are more similar to each other than to those in different clusters, enabling the discovery of natural groupings and hidden patterns in complex datasets.

Types of Clustering

1. Hard Clustering
2. Soft Clustering


1. Hard Clustering
In hard clustering, each data point strictly belongs to exactly one cluster, no overlap is allowed. This approach assigns a clear membership, making it easier to interpret and use for definitive segmentation tasks.

2. Soft Clustering
Soft clustering assigns each data point a probability or degree of membership to multiple clusters simultaneously, allowing data points to partially belong to several groups.
```
