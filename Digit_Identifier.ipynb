```python
# @title
from io import IncrementalNewlineDecoder
import sklearn
from sklearn.datasets import fetch_openml
import random
import numpy as np # Added numpy import
%matplotlib Inline
import matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
mnist = fetch_openml('mnist_784')
mnist
x, y =mnist['data'], mnist['target']
random_digit = x.iloc[random.randint(0, len(x)-1)] # Corrected line to select a random row
some_random_digit = random_digit.values.reshape(28,28) # Reshaping to 28x28, as MNIST images are 28x28
plt.imshow(some_random_digit, cmap = matplotlib.cm.binary)
x_train, x_test = x.iloc[:60000], x.iloc[60000:70000] # Corrected to use .iloc for row slicing
y_train, y_test = y.iloc[:60000], y.iloc[60000:70000] # Corrected to use .iloc for row slicing
shuffle_index = np.random.permutation(len(x_train)) # Adjusted permutation range to length of x_train
x_train, y_train = x_train.iloc[shuffle_index], y_train.iloc[shuffle_index] # Corrected to use .iloc for shuffling
model = LogisticRegression(tol=0.1, max_iter=2000) #tol is the tolerance for the stopping criteria
                                                    # ma-iter : maximum number of the interactions taken
model.fit(x_train, y_train)     #this step takes a lot of time
#finding the output
test_num = int(input('Select any number 1 and 10,000 to select the image'))
test_image_data =x_test.iloc[test_num] # Corrected to use .iloc for selecting by integer position
test_image  = test_image_data.values.reshape(28,28 )

precidition = model.predict([test_image_data])

plt.imshow(test_image, cmap = matplotlib.cm.binary)
plt.title(f'selected image number: {test_num}')
plt.xlabel(f'the omage has been classified as : {precidition[0 ]}')
```
